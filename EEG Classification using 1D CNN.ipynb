```python
from glob import glob #To read files from folder
import os
import mne
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

```python
#Read all files from the dataverse folder containing EEG recording for healthy and Schizophrenic patients
all_file_path=glob('dataverse_files/*.edf')
print(len(all_file_path))
```

```python
#To seperate the healthy patients files from the schizophrenic patients files
healthy_file_path=[i for i in all_file_path if 'h' in i.split('\\')[1]]
patient_file_path=[i for i in all_file_path if 's' in i.split('\\')[1]]
print(len(healthy_file_path),len(patient_file_path))
```

```python
#Function load data and Create 5secs overlapping epochs
def read_data(file_path):
    data=mne.io.read_raw_edf(file_path,preload=True) #use mne to read raw edf file
    data.set_eeg_reference() #Create a reference from the avg of all the channels
    data.filter(l_freq=0.5,h_freq=45) #Apply low-pass and high-pass filter to the continous data
    epochs=mne.make_fixed_length_epochs(data,duration=5,overlap=1) #Create epochs object from data
    array=epochs.get_data() #get epoch data as numpy array
    return array
```

```python
#Load the data, create epochs for all the files and store them in numpy array
control_epochs_array=[read_data(i) for i in healthy_file_path]
patient_epochs_array=[read_data(i) for i in patient_file_path]
```

```python
#Creating labels for the files
control_epochs_labels=[len(i)*[0] for i in control_epochs_array] #label all epochs in each control file zero
patient_epochs_labels=[len(i)*[1] for i in patient_epochs_array] #label all epochs in each patient file one
len(control_epochs_labels), len(patient_epochs_array)
```

```python
#We need to combine both data and the labels, to form a dataset for the NN
epochs_array=control_epochs_array+patient_epochs_array #one dataset from healthy and schizophrenic data
epochs_labels=control_epochs_labels+patient_epochs_labels #one dataset from healthy and schizophrenic labels
len(epochs_array), len(epochs_labels)
```

```python
#Grouping the data - Avoid grouping based on epochs or the machine learning model would have seen the unseen data, group based on participants
groups=[[i]*len(j) for i,j in enumerate(epochs_array)]
len(groups)
```

```python
#Reshape data because CNN expects data in a particular shape. First check shape by converting list to array
epochs_array=np.vstack(epochs_array)
epochs_labels=np.hstack(epochs_labels)
groups_array=np.hstack(groups)
```

```python
epochs_array.shape, epochs_labels.shape, groups_array.shape
```

Most deep learning frameworks like TensorFlow/Keras and PyTorch expect input data to follow specific conventions regarding the order of dimensions.

**-TensorFlow/Keras:** By default, they use the channels-last format, where the dimensions are arranged as **(batch_size, height, width, channels)** for 2D data or **(batch_size, time_steps, features)** for 1D data.
**-PyTorch:** By default, it uses the channels-first format, where the dimensions are arranged as **(batch_size, channels, height, width)** for 2D data or **(batch_size, channels, time_steps)** for 1D data.
Moving the channels to the end ensures compatibility with these frameworks when using their default configurations.

```python
#Move the channel no to end of the axis, because CNN(Keras) expects it at the end
epochs_array=np.moveaxis(epochs_array,1,2)
epochs_array.shape
```

```python
#Importing the CNN library
from tensorflow.keras.layers import Conv1D,BatchNormalization,LeakyReLU,MaxPool1D,\
GlobalAveragePooling1D,Dense,Dropout,AveragePooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.backend import clear_session
```

```python
#Creating the CNN Archictecture
def cnnmodel():
    clear_session()
    model=Sequential() #Using a sequential model for 1D
    model.add(Conv1D(filters=5,kernel_size=3,strides=1,input_shape=(1250,19))) #1
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2,strides=2))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1)) #2
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2,strides=2))
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1)) #3
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2,strides=2))
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1)) #4
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2,strides=2))
    model.add(Conv1D(filters=5,kernel_size=3,strides=1)) #5
    model.add(LeakyReLU())
    model.add(GlobalAveragePooling1D())
    model.add(Dense(1,activation='sigmoid'))

    model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model=cnnmodel()
model.summary()
```

```python
#Create 5 Fold cross-validation loop and train the model for each loop
from sklearn.model_selection import GroupKFold, LeaveOneGroupOut
from sklearn.preprocessing import StandardScaler
gkf=GroupKFold()
```

```python
#Splitting the train and val data for each group
accuracy=[]
for train_index,val_index in gkf.split(epochs_array,epochs_labels,groups=groups_array):
    train_features,train_labels=epochs_array[train_index],epochs_labels[train_index]
    val_features,val_labels=epochs_array[val_index],epochs_labels[val_index]
    scaler=StandardScaler()
    train_features=scaler.fit_transform(train_features.reshape(-1,train_features.shape[-1])).reshape(train_features.shape)
    val_features=scaler.fit_transform(val_features.reshape(-1,val_features.shape[-1])).reshape(val_features.shape)
    model=cnnmodel()
    model.fit(train_features,train_labels,epochs=10, batch_size=100,validation_data=(val_features,val_labels))
    accuracy.append(model.evaluate(val_features,val_labels)[1])
    break
```
